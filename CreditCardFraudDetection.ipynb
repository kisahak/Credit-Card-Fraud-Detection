## Model Prediction: Isolation Forest Algorithm

**Identifying Credit Card Fraud: The Concept of Finding Outliers**
Credit card fraud detection is essentially an outlier detection problem. Frauds are rare events - anomalies that stand out from the usual pattern of transactions.

**Why Isolation Forest?**
An effective way to detect these anomalies is by using the Isolation Forest Algorithm. This approach is particularly well-suited for datasets with a high degree of imbalance, like our credit card transaction dataset.

**Understanding the Isolation Forest Algorithm:**
- *The Core Idea*: At its heart, the Isolation Forest algorithm is similar to a classic decision tree. However, instead of making splits to classify data, it isolates anomalies.

- *Depth as a Score*: The key metric in this algorithm is the depth of the leaf nodes. We can think of this depth as a score.
- *Interpreting the Score*: A larger score, indicating greater depth, correlates with the data point being an anomaly. Essentially, the deeper the leaf node where a sample ends up, the more anomalous the sample is considered. These data points are the ones that are farthest away from the densest area of the data. In other words, the outliers of the dataset! 
- *Anomaly Detection*: The Isolation Forest algorithm looks at the depth of leaf nodes and considers deeper nodes as more likely to be anomalies. This is based on the principle that anomalies are few and different, and thus easier to isolate. 



```python
import sklearn
import scipy
from sklearn.metrics import classification_report,accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from pylab import rcParams
rcParams['figure.figsize'] = 14, 8
RANDOM_SEED = 42
LABELS = ["Legit", "Fraud"]
```

```python
outlier_fraction = len(fraud)/float(len(legit))

classifiers = {
    "Isolation Forest":IsolationForest(n_estimators=100, max_samples=len(X), 
                                       contamination=outlier_fraction,random_state=state, verbose=0),
    #we can add more classifiers to pipe line here later 
}
```

```python
n_outliers = len(fraud)
for i, (clf_name,clf) in enumerate(classifiers.items()):
    #Fit the data and tag outliers
    if clf_name == "Isolation Forest":
        clf.fit(X)
        scores_prediction = clf.decision_function(X)
        y_pred = clf.predict(X)
    #Reshape the prediction values to 0 for Valid transactions , 1 for Fraud transactions
    y_pred[y_pred == 1] = 0
    y_pred[y_pred == -1] = 1
    n_errors = (y_pred != Y).sum()
    # Run Classification Metrics
    print("{}: {}".format(clf_name,n_errors))
    print("Accuracy Score :")
    print(accuracy_score(Y,y_pred))
    print("Classification Report :")
    print(classification_report(Y,y_pred))
```
